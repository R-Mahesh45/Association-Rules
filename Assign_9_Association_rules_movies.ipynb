```python
import pandas as pd
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder   ## out data is encoded already
```

```python
movie = pd.read_csv("/content/my_movies.csv")
```

```python
movie.head()
```

```python
data = movie.iloc[:, 5:]  ## i have sorteed the movies which are watched by the IDs
data
```

```python
data.head()
```

```python
freq_itm = apriori(data, min_support=0.1, use_colnames=True)

freq_itm
```

```python
rules = association_rules(freq_itm, metric="lift", min_threshold=0.7)  ## min_threshold is confidence
rules
```

```python
rules.sort_values('lift', ascending = False)[0:20]
```

```python
rules[rules.lift>2]
```

```python
rules.sort_values('confidence', ascending = False)[0:20]
```

```python
rules.sort_values('support', ascending = False)[0:20]
```
